```python
from math import log, sqrt
import numpy as np

from bokeh.plotting import figure, show
from bokeh.sampledata.antibiotics import data as df
from bokeh.io import output_notebook 
from bokeh.models import Arrow, NormalHead, OpenHead, VeeHead
from bokeh.models import Span




output_notebook()
width = 500
height = 500

p = figure(width=width, height=height, title="",
  #  x_axis_type=None, y_axis_type=None,
    x_range=(-2, 2), y_range=(-2, 2),
    min_border=0, outline_line_color="black")
 #   background_fill_color="#f0e1d2")

p.xgrid.grid_line_color = None
p.ygrid.grid_line_color = None


# circular axes and lables
p.circle(0, 0, radius=1.0, fill_color=None, line_color="black")
p.circle(0, 0, radius=2.0, fill_color=None, line_color="black")

xax = Span(location=0,dimension='width', level='glyph')
p.add_layout(xax)

#yaxis
yax = Span(location=0,dimension='height', level='glyph')
p.add_layout(yax)

angle = 3 * np.pi / 8.0 
x = np.cos(angle)
y = np.sin(angle)

vh = VeeHead(size=4, fill_color="black")
p.add_layout(Arrow(end=vh, x_start=0.0, y_start=0.0, x_end=0.0, y_end=1.0, line_width=2))



show(p)
```
